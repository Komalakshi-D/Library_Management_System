```python
# Convert 'LoanDate' column to datetime format
loans_data['LoanDate'] = pd.to_datetime(loans_data['LoanDate'])

# Extract month from 'LoanDate' and store it in a new column 'Month'
loans_data['Month'] = loans_data['LoanDate'].dt.to_period('M')

# Set up the plot figure with a specific size
plt.figure(figsize=(12, 6))

# Count the number of loans per month, sort by index (month), and plot as a line graph
loans_data['Month'].value_counts().sort_index().plot(kind='line')

# Set the title, xlabel, and ylabel for the plot
plt.title('Number of Loans per Month')
plt.xlabel('Month')
plt.ylabel('Number of Loans')

# Display the plot
plt.show()
```

```python
# Convert 'MembershipDate' column to datetime format
members_data['MembershipDate'] = pd.to_datetime(members_data['MembershipDate'])

# Extract year from 'MembershipDate' and store it in a new column 'Year'
members_data['Year'] = members_data['MembershipDate'].dt.year

# Set up the plot figure with a specific size
plt.figure(figsize=(12, 6))

# Count the number of members for each year, sort by index (year), and plot as a pie chart
members_data['Year'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%')

# Set the title for the plot
plt.title('Distribution of Members by Membership Year')

# Display the plot
plt.show()
```

```python
# Merge books and reviews data to include ratings
books_reviews = pd.merge(books_data, reviews_data, on='BookID')

# Aggregate data by rating and calculate average price
avg_price_by_rating = books_reviews.groupby('Rating')['Price'].mean().reset_index()

# Plot the relationship between rating and average price using a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Rating', y='Price', data=avg_price_by_rating, marker='o', color='blue')
plt.title('Relationship between Rating and Average Price of Books')
plt.xlabel('Rating')
plt.ylabel('Average Price')
# Add grid lines for better readability
plt.grid(True)
plt.show()
```

```python
# Query to find members who returned books after the due date and calculate days late
query = """
SELECT L.MemberID, (L.ReturnDate - L.DueDate) as DaysLate
FROM Loans L
WHERE L.ReturnDate > L.DueDate
"""

# Execute the query
cursor.execute(query)
data = cursor.fetchall()

# Extract member IDs and days late from the data
member_ids = [row[0] for row in data]
days_late = [row[1] for row in data]

# Create a scatter plot of days late vs. members who returned the book
plt.figure(figsize=(10, 5))
plt.bar(member_ids, days_late, color='red')
plt.xlabel('Member ID')
plt.ylabel('Days Late')
plt.title('Days Late in Returning Books vs. Members')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Query to count the number of times each book has been borrowed
query = """
SELECT B.Title, COUNT(L.BookID) as NumberOfBorrows
FROM Loans L
JOIN Books B ON L.BookID = B.BookID
GROUP BY B.Title
ORDER BY NumberOfBorrows DESC
LIMIT 10  -- Limiting to top 10 most read books
"""

# Execute the query
cursor.execute(query)
data = cursor.fetchall()

# Extract book titles and number of borrows
book_titles = [row[0] for row in data]
number_of_borrows = [row[1] for row in data]

# Create a bar chart of the most read books
plt.figure(figsize=(10, 6))
plt.bar(book_titles, number_of_borrows, color='purple')
plt.xlabel('Book Titles')
plt.ylabel('Number of Borrows')
plt.title('Top 10 Most Read Books in the Library')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Query to count the number of books written by authors from each country
query = """
SELECT Nationality, COUNT(*) as NumberOfBooks
FROM Authors
JOIN Books ON Authors.AuthorID = Books.AuthorID
GROUP BY Nationality
ORDER BY NumberOfBooks DESC
"""

# Execute the query
cursor.execute(query)
data = cursor.fetchall()

# Extract country nationalities and number of books
countries = [row[0] for row in data]
number_of_books = [row[1] for row in data]

# Create a bar chart of the number of books written by authors from each country
plt.figure(figsize=(12, 6))
plt.bar(countries, number_of_books, color='green')
plt.xlabel('Country')
plt.ylabel('Number of Books')
plt.title('Number of Books Written by Authors from Each Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Query to count on-time returns, early returns, and late returns
query = """
SELECT
    SUM(CASE WHEN ReturnDate < DueDate THEN 1 ELSE 0 END) as EarlyReturns,
    SUM(CASE WHEN ReturnDate = DueDate THEN 1 ELSE 0 END) as OnTimeReturns,
    SUM(CASE WHEN ReturnDate > DueDate THEN 1 ELSE 0 END) as LateReturns
FROM Loans
"""

# Execute the query
cursor.execute(query)
result = cursor.fetchone()

# Extract the counts
early_returns = result[0]
on_time_returns = result[1]
late_returns = result[2]

# Data for plotting
labels = ['Early Returns', 'On Time Returns', 'Late Returns']
sizes = [early_returns, on_time_returns, late_returns]
colors = ['green', 'blue', 'red']
explode = (0.1, 0, 0)  # explode the first slice (early returns)

# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title('Percentage of Book Returns: Early, On Time, and Late')
plt.show()
```

```python
# Query to count the number of books in each category
query = """
SELECT C.CategoryName, COUNT(*) as NumberOfBooks
FROM Categories C
JOIN Books B ON C.CategoryID = B.CategoryID
GROUP BY C.CategoryName
ORDER BY NumberOfBooks DESC
LIMIT 15  -- Limiting to top 15 categories
"""

# Execute the query
cursor.execute(query)
data = cursor.fetchall()

# Extract category names and number of books
categories = [row[0] for row in data]
number_of_books = [row[1] for row in data]
# Create a bar chart of the top 10 categories of books read
plt.figure(figsize=(10, 6))
plt.bar(categories, number_of_books, color='olive')
plt.xlabel('Category')
plt.ylabel('Number of Books')
plt.title('Top 15 Categories of Books Read')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Query to get the number of books published by each publisher
query = """
SELECT P.Name, COUNT(B.BookID) as NumberOfBooks
FROM Publishers P
JOIN Books B ON P.PublisherID = B.PublisherID
GROUP BY P.Name
ORDER BY NumberOfBooks DESC
"""

# Execute the query
cursor.execute(query)
data = cursor.fetchall()

# Extract publisher names and number of books published
publishers = [row[0] for row in data]
number_of_books = [row[1] for row in data]

# Create a bar chart of the number of books published by each publisher
plt.figure(figsize=(12, 8))
plt.bar(publishers, number_of_books, color='plum')
plt.xlabel('Publisher')
plt.ylabel('Number of Books Published')
plt.title('Number of Books Published by Each Publisher')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Query to get the necessary data
query = """
SELECT B.Title, B.Price, R.Rating, P.Name as Publisher
FROM Books B
JOIN Reviews R ON B.BookID = R.BookID
JOIN Publishers P ON B.PublisherID = P.PublisherID
WHERE B.Price IS NOT NULL AND R.Rating IS NOT NULL
"""

# Execute the query
cursor.execute(query)
data = cursor.fetchall()

# Load data into a pandas DataFrame
columns = ['Title', 'Price', 'Rating', 'Publisher']
df = pd.DataFrame(data, columns=columns)

# Filter the DataFrame for low price and high rating
low_price_threshold = 500  # Example threshold for low price
high_rating_threshold = 4.5  # Example threshold for high rating

filtered_df = df[(df['Price'] < low_price_threshold) & (df['Rating'] >= high_rating_threshold)]

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Price'], filtered_df['Rating'], c='blue', marker='o')
plt.title('Books with Low Prices and High Ratings')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.grid(True)

# Annotate the plot with book titles
for i, row in filtered_df.iterrows():
    plt.annotate(row['Title'], (row['Price'], row['Rating']))

# Show the plot
plt.show()
```

```python
# Query to get the count of duplicate titles
query_duplicates = """
SELECT COUNT(*) AS DuplicateCount
FROM (
    SELECT Title
    FROM Books
    GROUP BY Title
    HAVING COUNT(*) > 1
) AS DupTitles
"""

# Query to get the count of unique titles
query_unique = """
SELECT COUNT(*) AS UniqueCount
FROM (
    SELECT Title
    FROM Books
    GROUP BY Title
    HAVING COUNT(*) = 1
) AS UniqueTitles
"""

# Execute the queries
cursor.execute(query_duplicates)
duplicate_count = cursor.fetchone()[0]

cursor.execute(query_unique)
unique_count = cursor.fetchone()[0]

# Data for the pie chart
labels = ['Repeated Titles', 'Unique Titles']
sizes = [duplicate_count, unique_count]

# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Repeated Titles over Unique Titles')
plt.show()
```

```python
# Run the SQL query and load data into a DataFrame
query = """
SELECT Books.BookID, Books.Title, COUNT(Loans.LoanID) AS LoanCount
FROM Books
LEFT JOIN Loans ON Books.BookID = Loans.BookID
GROUP BY Books.BookID, Books.Title;
"""

# Execute the query
cursor.execute(query)
data = cursor.fetchall()

# Classify books
def classify_books(loan_count):
    if loan_count == 0:
        return 'Not Read'
    elif loan_count <= 1:
        return 'Less Read'
    else:
        return 'Most Read'

df['Category'] = df['LoanCount'].apply(classify_books)

# Plot the distribution
category_counts = df['Category'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='pie', autopct='%1.1f%%', color=['orange', 'blue', 'green'])
plt.title('Distribution of Books Based on Number of Times Borrowed')
plt.show()
```
